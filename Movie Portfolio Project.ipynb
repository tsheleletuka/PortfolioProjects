```python
# Using factorize - this assigns a random numeric value for each unique catergoical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
df
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can take a look at the one that have a high correlation ( > 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
df_numerised = df

# Iterate through each column in the DataFrame
for col_name in df_numerised.columns:
    # Check if the data type of the column is 'object' (typically indicates categorical data)
    if(df_numerised[col_name].dtype == 'object'):
        # Convert the column to the 'category' data type
        df_numerised[col_name] = df_numerised[col_name].astype('category')
        # Replace the categorical values with their corresponding codes
        df_numerised[col_name] = df_numerised[col_name].cat.codes
        
df_numerised
```

```python
df_numerised.corr(method='pearson')
```

```python
correlation_matrix = df_numerised.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
